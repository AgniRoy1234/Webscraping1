```
import requests
from bs4 import BeautifulSoup as BS
import pandas as pd
import time
```

```
url="https://wistia.com/about/yearbook"
```

```
r=requests.get(url)
```

```
soup=BS(r.content,"lxml")
```

```
a=soup.findAll("li",class_="jsx-2483422208 Wistian")
```

```
info=[]
for i in a:
  info.append([i.find("p",class_="jsx-4277224649 Wistian__name").text,i.find("p",class_="jsx-4277224649 Wistian__team").text])
```
