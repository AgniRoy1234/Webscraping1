```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
```

```python
header={"User-Agent":"Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/80.0.3987.149 Safari/537.36"}
```

```python
r=requests.get("https://www.finology.in/team.aspx",headers=header)
```

```python
soup=BeautifulSoup(r.content,"lxml")
```

```python
soup.findAll("div",{"class":"card border-0 shadow teamcard"})
```

```python
info=[]
for x in soup.findAll("div",{"class":"card border-0 shadow teamcard"}):
    y=x.find("h5",{"class":"card-title mb-0"})
    info.append((y.text,y.find("span").text))
    
```

```python
info
```

```python
def funct1(tup):
    l=[[],[]]
    k=tup[0].index(tup[1])
    l[0]=tup[0][:k].strip()
    l[1]=tup[1].strip()
    return l
```

```python
info1=list(map(funct1,info))
```

```python
info1
```

```python
info1=info1[:-1]
```

```python
info1
```

```python
bio=[]
soup.findAll("div",{"class":"card-text text-black-50"})
for i in soup.findAll("div",{"class":"card-text text-black-50"}):
    bio.append(i.text.strip())
```

```python
bio
```

```python
bio=bio[:-1]
```

```python
final_info=[]
for i in range(0,len(bio)):
    info1[i].append(bio[i])
    
```
